```python
df = pd.read_csv(url, names=names)
data = df
print(data.corr())
```

What column has the highest correlation with class? Find the mean() value of that column

```python
# YOUR CODE HERE
```

```python
correlation_with_class = df.corr()['class'].abs().sort_values(ascending=False)

# Exclude 'class' itself from the list
highest_corr_column = correlation_with_class.index[1]
print(f"Column with the highest correlation with 'class': {highest_corr_column}")

mean_of_highest_corr_column = df[highest_corr_column].mean()
print(f"Mean value of the '{highest_corr_column}' column: {mean_of_highest_corr_column:.2f}")
```

Create a new dataframe with only people that are above the mean found in the above cell.

```python
# YOUR CODE HERE
```

```python
# Assuming 'plas' was the column with the highest correlation
# If it was different, you would use the variable `highest_corr_column` from the previous cell.
df_above_mean = df[df['plas'] > df['plas'].mean()]
display(df_above_mean.head())
```

Create a new dataframe that only includes people who are above the average value for each column in the dataset. In other words, find the average value for each column, and then find which users have values above all of these average column values.

```python
# YOUR CODE HERE
```

```python
df_means = df.mean()
print(df_means.shape)
# Create a boolean DataFrame where each value is True if it's above its column mean
df_above_all_means = df[ (df >= df_means).all(axis=1) ]
display(df_above_all_means.head())
```
